### Write a Python programming to display a bar chart of the popularity of programming Languages

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp

language = ['Java', 'Python', 'PHP', 'JavaScript', 'C', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(language,popularity, color = ['pink', 'lightpink','salmon', 'lightcoral','tomato', 'maroon'])
plt.xlabel('\nProgramming Languages')
plt.ylabel('Popularity as Percentage')
plt.title('How popular is your programming language?\n')
```
