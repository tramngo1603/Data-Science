### Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016


```python
# Make a dictionary of given data 

Alphabet_dict = [('10-03-16', [774.25,776.065002,769.5,772.559998]), 
                    ('10-04-16', [776.030029,778.710022,772.890015,776.429993]),
                    ('10-05-16', [779.309998,782.070007,775.650024,776.469971]),
                    ('10-06-16', [779,780.47998,775.539978,776.859985]),
                    ('10-07-16', [779.659973,779.659973,770.75,775.080017])]


# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Plot a line 
x = []
y_open = []
y_low = []
y_high = []
y_close = []

for i in range(len(Alphabet_dict)):
    date = Alphabet_dict[i][0]
    x.append(date)
for i in range(len(Alphabet_dict)):
    open_ = Alphabet_dict[i][1][0]
    y_open.append(open_)
for i in range(len(Alphabet_dict)):
    low_ = Alphabet_dict[i][1][2]
    y_low.append(low_)
for i in range(len(Alphabet_dict)):
    high_ = Alphabet_dict[i][1][1]
    y_high.append(high_)
for i in range(len(Alphabet_dict)):
    close_= Alphabet_dict[i][1][3]
    y_close.append(close_)
    

plt.plot(x,y_open,label = 'Open', c = 'red') 
plt.plot(x,y_low, label = 'Low', c = 'blue')
plt.plot(x, y_high, label = 'High', c = 'orange') 
plt.plot(x, y_close, label = 'Close', c = 'green')

plt.legend()
plt.ylabel('Stock Price')
plt.xlabel('\nTime period from Oct 03,2016 to Oct 07, 2016')
plt.title('Alphabet Inc\'s Financial Data\n')
plt.show()
```
